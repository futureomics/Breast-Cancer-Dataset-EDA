```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

# Load Breast Cancer dataset
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer(as_frame=True)
cancer_df = cancer_data.frame

# Add target labels to the DataFrame
cancer_df['target'] = cancer_df['target'].map({0: 'malignant', 1: 'benign'})

# Set global Seaborn style
sns.set(style="whitegrid")

# Display the first few rows of the dataset
cancer_df.head()
```

Since the dataset has a large number of features, we can choose a few important ones for visualization. Let’s select some key features such as the mean of the radius, texture, and smoothness for the pairplot.

```python
# Pairplot for a subset of features
subset_features = ['mean radius', 'mean texture', 'mean smoothness', 'target']
sns.pairplot(cancer_df[subset_features], hue='target', palette='Set1', markers=["o", "s"])
plt.show()
```

This will create a pairplot for the selected features, showing relationships between them and color-coding by target (benign or malignant).



```python
# Boxplots for selected features across target classes (malignant vs benign)
features = ['mean radius', 'mean texture', 'mean smoothness', 'mean area']
plt.figure(figsize=(14, 10))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='target', y=feature, data=cancer_df, palette="Set2")
    plt.title(f'{feature} by Target')

plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(14, 12))
corr_matrix = cancer_df.iloc[:, :-1].corr()  # Exclude the target column for correlation
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={'shrink': .8})
plt.title('Feature Correlation Heatmap')
plt.show()
```

This will display a correlation matrix heatmap showing how the features are correlated with each other.

```python
# Count plot for target distribution (benign vs malignant)
plt.figure(figsize=(7, 5))
sns.countplot(x='target', data=cancer_df, palette="coolwarm")
plt.title('Benign vs Malignant Tumor Distribution')
plt.show()
```

```python
# Interactive scatter matrix using Plotly for selected features
fig = px.scatter_matrix(cancer_df, dimensions=['mean radius', 'mean texture', 'mean smoothness'],
                        color="target",
                        title="Interactive Scatter Matrix",
                        labels={col: col.replace('_', ' ') for col in ['mean radius', 'mean texture', 'mean smoothness']},
                        height=700)
fig.show()
```

```python
# Interactive 3D scatter plot using Plotly
fig = px.scatter_3d(cancer_df, x='mean radius', y='mean texture', z='mean smoothness',
                    color='target', symbol='target',
                    title="3D Scatter Plot of Mean Features")
fig.show()
```

Summary of Visualization Components:
Distribution Plots: Shows the distribution of the most relevant features in the dataset.
Pairwise Relationships: Pairplots visualize the relationships between key features, colored by tumor type (benign or malignant).
Boxplots: Useful for comparing the distribution of key features across tumor types.
Correlation Heatmap: Displays how different features correlate with each other.
Target Distribution: Shows the balance of benign and malignant tumors.
Interactive Plots: Using Plotly, you can create both a scatter matrix and a 3D scatter plot to interactively explore relationships between features.
